# Problem 1

What is the probability that at least 9 out of a group of 10 people who have been infected by a serious disease will survive, if the survival probability for the disease is 70%?

X = number of people survived

n = 10

p = 0.7

X can be modelled by a binomial distribution.

X ~ Bin(10, 0.7)

Required probability = P(X>=9) = P(X = 9) + P(X = 10)

= $[C(10,9) \times (0.7)^{9} \times (1 - 0.7)^{1}] + [C(10,10) \times (0.7)^{10} \times (1 - 0.7)^{0}] = 0.1493$
